```
df = pd.DataFrame()
df['rand'] = np.random.rand(100)
df['rand_sq'] = df['rand'] ** 2
df['rand_shift'] = df['rand'] + 2
```

```
df['counts_sq'] = df.index ** 2 
df['counts_sqrt'] = np.sqrt(df.index)
```

```
plt.plot(df['rand'])
plt.show()
```

```
plt.plot(df['rand'], color='purple')
plt.ylim([-0.1, 1.1])
plt.ylabel('Değerler')
plt.title('Rastgele Sayılar')
plt.show()
```

```
plt.plot(df['rand'], color='purple')
plt.plot(df['rand_shift'], color='green')
plt.ylim([-0.1, 3.1])
plt.ylabel('Değerler')
plt.title('Rastgele Diziler')
plt.show()
```

```
plt.scatter(x=df['rand'], y=df['rand_sq'])
plt.show()
```

```
plt.scatter(
    x=df['rand'],
    y=df['rand_sq'],
    color='purple',
    marker='+', s=10
)
plt.scatter(
    x=df['rand'],
    y=df['rand_shift'],
    color='green',
    marker='x', s=1
)
plt.show()
```

```
x = np.random.normal(10, 5, 1000)

plt.hist(x) 
plt.show()
```

```
x = np.random.normal(10, 5, 1000)

plt.hist(x, bins=40, color='red')
plt.title('Kutu Seçimi')
plt.xlabel('Rastgele Değerler')
plt.show()
```

```
x = np.random.normal(10, 5, 1000)
y = np.random.normal(15, 5, 10000)

plt.hist(x, color='blue', bins=np.arange(-10, 40), alpha=.5) 
plt.hist(y, color='red', bins=np.arange(-10, 40), alpha=.5)
plt.title('Manuel Kutu Yerleşimi')
plt.xlabel('Rastgele Değerler')

plt.show()
```

Çubuk grafikleri özellikle, kategorik (yani sınırlı sayıda farklı değer içeren) değişkenleri incelemek için yararlı araçlardır. Bu ödevde, kategorik bir değişken içeren bir veri kümesi bulmanız gerekecek. Bunun için Kaggle sitesindeki veri setlerine bir göz atabilirsiniz. Daha sonra, bulduğunuz veri kümesini Pandas'ı kullanarak bir veri çerçevesine aktarın. Veri kümenizde yer alan kategorik değişkenlerden iki tanesini seçin ve subplot kullanarak, bu iki değişkenin çubuk grafiklerini çıkarın. Bulgularınızı mentörünüzle tartışmayı unutmayın!
